# 3. A description of final product.

This infographic communicates the unequal distribution of Name, Image, and Likeness (NIL) value across both conferences and individual players in college football. The donut chart highlights how NIL money is heavily concentrated within the Power 4 conferences, with the SEC alone generating over seven times more NIL value than all non-power conferences combined. This demonstrates that financial opportunity in college sports is strongly tied to conference power, visibility, and resources.

The Lorenz curve further illustrates inequality at the player level, showing that most NIL value is concentrated among a very small group of elite athletes, producing a Gini coefficient of 0.215. This shape indicates a market where star power, rather than broad participation, drives compensation. The bar chart reinforces this pattern by comparing the top four NIL earners to the bottom sixty, revealing that a handful of players earn more than the majority combined.

Together, these visualizations show that NIL compensation is not evenly distributed and often reinforces existing power structures within college football. Instead of creating equal opportunity, NIL currently functions as a winner-take-most system, where both conference affiliation and player status significantly determine financial outcomes.
